```python
# CONVERTING THE CATEGORICAL DATA TO NUMERICAL VALUES AND DROPPING THE 'NO' COLUMNS AT THE SAME TIME
market_segmentation_data = pd.get_dummies(df, columns=['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting', 'Gender'], drop_first=True)
```

```python
market_segmentation_data.head(3)
```

```python
# WORKING ON THE COLUMN 'LIKE' & 'VISITFREQUENCY', THAT INCLUDES STRING VALUES TOO
market_segmentation_data['Like'].unique()
```

```python
# ALTERNATE
# market_segmentation_data[market_segmentation_data['Like'] == 'I love it!+5']['Like'] = int(+5)

# REPLACING THE STRING VALUE TO INTEGERS
market_segmentation_data['Like'] = market_segmentation_data['Like'].replace({'I love it!+5' : int(+5), 'I hate it!-5' : int(-5)})
```

```python
# CONVERTING ALL VALUES FROM TYE STR TO INT
market_segmentation_data['Like'] = market_segmentation_data['Like'].astype(int)
```

```python
# CHECKING UNIQUE VALUES IN THE 'VISITFREQUENCY' COLUMN
market_segmentation_data['VisitFrequency'].unique()
```

```python
market_segmentation_data['VisitFrequency'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
feature_array = enc.fit_transform(market_segmentation_data[['VisitFrequency']]).toarray()
```

```python
enc.categories_
```

```python
feature_labels = np.array(enc.categories_).ravel()
features = pd.DataFrame(feature_array, columns=feature_labels)
features
```

```python
market_segmentation_data = pd.concat([market_segmentation_data, features], axis=1)
```

```python
market_segmentation_data
```

```python
# DROPPING THE 'VISITFREQUENCY' COLUMN AS WE HAVE ENLISTED ITS VALUES AS COLUMN LABELS
market_segmentation_data.drop(['VisitFrequency'], axis=1, inplace=True)
```

```python
market_segmentation_data.columns
```

```python
# RENAMING THE COLUMNS
market_segmentation_data.rename(columns={'yummy_Yes':'Yummy', 'convenient_Yes':'Convenient', 
       'spicy_Yes':'Spicy', 'fattening_Yes':'Fattening', 'greasy_Yes':'Greasy', 'fast_Yes':'Fast', 'cheap_Yes':'Cheap',
       'tasty_Yes':'Tasty', 'expensive_Yes':'Expensive', 'healthy_Yes':'Healthy', 'disgusting_Yes':'Disgusting',
       'Gender_Male':'Gender'}, inplace=True)

market_segmentation_data.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = market_segmentation_data[['Age', 'Yummy', 'Convenient', 'Spicy', 'Fattening', 'Greasy',
       'Fast', 'Cheap', 'Tasty', 'Expensive', 'Healthy', 'Disgusting',
       'Gender', 'Every three months', 'More than once a week', 'Never',
       'Once a month', 'Once a week', 'Once a year']]

y = market_segmentation_data['Like']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.displot((y_test - predictions), bins=30, kde=True)
```

```python
from sklearn.svm import SVC

model= SVC()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1, 1, 10, 100, 1000], 'gamma':[1, 0.1, 0.01, 0.001, 0.0001]}
```

```python
grid = GridSearchCV(SVC(), param_grid, verbose=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```
